```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
pd1  = pd.read_csv(r'E:\Desktop\机器学习\具体计算\Be-Fcc-ML-Data\Be-Fcc-AllData.csv',)
X = pd1.iloc[:,0:2]
Y = pd1.iloc[:,-1]

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y,test_size=0.3)
```

```python
clf = DecisionTreeRegressor(random_state=0)
rfc = RandomForestRegressor(random_state=0)
```

```python
clf.fit(Xtrain,Ytrain)
rfc.fit(Xtrain,Ytrain)

score_c = clf.score(Xtest,Ytest)
score_r = rfc.score(Xtest,Ytest)

print(f'Single Tree:{score_c}',
        f'Random Forest:{score_r}'
    )
```

```python
clf = DecisionTreeRegressor()
clf_s = cross_val_score(clf,X,Y,cv=10)

rfc = RandomForestRegressor(n_estimators=25)
rfc_s = cross_val_score(rfc,X,Y,cv=10)

plt.plot(range(1,11),rfc_s,label = "RandomForest")
plt.plot(range(1,11),clf_s,label = "Decision Tree")
plt.legend()
plt.show()
```

```python
rfc_l = []
clf_l = []
for i in range(10):
    rfc = RandomForestRegressor(n_estimators=25)
    rfc_s = cross_val_score(rfc,X,Y,cv=10,scoring="neg_mean_squared_error").mean()
    rfc_l.append(rfc_s)
    clf = DecisionTreeRegressor()
    clf_s = cross_val_score(clf,X,Y,cv=10,scoring="neg_mean_squared_error").mean()
    clf_l.append(clf_s)

plt.plot(range(1,11),rfc_l,label = "Random Forest")
plt.plot(range(1,11),clf_l,label = "Decision Tree")
plt.legend()
plt.show()
```
