```python
#特征化一个pymatgen结构
from matminer.featurizers.base import MultipleFeaturizer
#元素组分特征
from matminer.featurizers.composition import ElementProperty, Stoichiometry
#结构相关特征：化学序（当前结构和完全随机结构的差别）
from matminer.featurizers.structure import ChemicalOrdering,StructureComposition

from pymatgen.ext.matproj import MPRester
#以著名的Heusler金属间化物 MnAlCu2 MnCu2Sn为例
#在浏览器上从MP可以查到mpid分别为:mp-3574 mp-22221
#复习一下使用mp获取结构
with MPRester(key) as m:
    structure = m.get_structure_by_material_id("mp-3574")
    structure2 = m.get_structure_by_material_id("mp-22221")
    print(structure)
    print('----------')
    print(structure2)

#可以把多个特征的生成器组合成一个总的特征生成器
#除化学计量比外，元素特征用magpie预设的，具体有哪些请翻阅官方文档和代码
featurizer = MultipleFeaturizer([
    ChemicalOrdering(),
    StructureComposition(Stoichiometry()),
    StructureComposition(ElementProperty.from_preset("magpie")),
])

#特征化输入结构，得到描述符X
X=featurizer.featurize(structure)
X2=featurizer.featurize(structure2)
print('----------')
print(X)
print('----------')
print(X2)
#也可以同时特征化多个结构(还会显示进度条)
strulist=[structure,structure2]
X_all=featurizer.featurize_many(strulist)
```
