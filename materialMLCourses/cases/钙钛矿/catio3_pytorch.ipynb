```python
import csv
import os

import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class CaTiO3Dataset(Dataset):
    def __init__(self, filepath):
        id_prop_file = os.path.join(filepath)
        with open(id_prop_file) as f:
            reader = csv.reader(f)
            self.id_prop_data = [row for row in reader]

    def __getitem__(self, index):
        id = self.id_prop_data[index][0]
        x = torch.tensor(list(map(float, self.id_prop_data[index][2:])))
        y = torch.tensor(float(self.id_prop_data[index][1]))
        return id, x, y

    def __len__(self):
        return len(self.id_prop_data)
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1 = torch.nn.Linear(4, 2)
        self.linear2 = torch.nn.Linear(2, 1)
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
        x = self.sigmoid(self.linear1(x))
        x = self.sigmoid(self.linear2(x))
        return x
```

```python
# 实例化数据集
dataset = CaTiO3Dataset('./catio3.csv')
train_loader = DataLoader(dataset=dataset, batch_size=32, shuffle=True, num_workers=2)
# 实例化模型

model = Model()

# 实例化criterion
criterion = torch.nn.MSELoss()

# 实例化优化器
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
# 训练

for epoch in range(30):
    ids = []
    tagets = []
    preds = []
    for i, data in enumerate(train_loader, 0):
        id, inputs, labels = data
        print(id)
        y_pred = model(inputs)
        loss = criterion(y_pred, labels)
        print('epoch:{}, loss:{}'.format(epoch, loss.item()))

        ids += id
        tagets += labels.view(-1).tolist()
        preds += y_pred.view(-1).tolist()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
with open('results.csv', 'w') as f:
    writer = csv.writer(f)
    for id, target, pred in zip(ids, tagets, preds):
        writer.writerow((id, target, pred))

df = pd.read_csv('results.csv', header=None, names=['ID', 'Target', 'Prediction'])
df.to_csv('results.csv', index=False)
```
